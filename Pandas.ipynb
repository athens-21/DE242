# **Pandas Series**

**Pandas Series** - Merging Deries


*   concat()




```python
s5 = pd.Series(['a', 'b'], index=[1, 2])
s6 = pd.Series(['c', 'd'], index=[3, 4])
s5s6 = pd.concat([s5, s6])
s5s6
```

**Pandas Series** - Data Deletion


*   drop()



```python
s7 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
s7.drop('a')
```



*   Drop by range of indics



```python
s7drop2 = s7.drop(s7[1:2].index)
s7drop2
```





*   Drop by Boolean condition





```python
s7drop3 = s7[s7>3]
s7drop3
```

**Pandas Series** - Operation on data

```python
s9 = pd.Series([-1, -2, -3], index=[0, 1, 2])
s9
```

Using concat( ) function

Appending Column(s)

```python
df1 = pd.DataFrame({'col1':[1, 2, 3]}, index = [0, 1, 2])
df2 = pd.DataFrame({'col2':[5, 6, 7]}, index = [2, 3, 4])
df3 = pd.concat([df1, df2], axis=1)
df3.head()
```

Appending Row(s)

```python
df4 = pd.DataFrame({'col1':9, 'col2':10}, index=[3])
df5 = pd.concat([df3, df4], axis=0)
df5
```

Dropping Row(s)

```python
df1 = pd.DataFrame({'col1':[1, 2, 3]}, index = [0, 1, 2])
df2 = pd.DataFrame({'col2':[5, 6, 7]}, index = [2, 3, 4])
df3 = pd.concat([df1.reset_index(drop=True), df2.reset_index(drop=True)], axis=1)
df3
```

```python
df5 =df3.drop(df3.index[0:2], inplace=False)
df5
```

***df[begin:end:step]*** to select zero or more rows

*   **‘begin’** and **‘end’** can be integer index or labeled index and their
values MUST BE provided

*   **‘step’** is optional but if provided it MUST BE a number.
