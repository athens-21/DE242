```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = np.random.randn(100)
plt.hist(x=data, bins=10, color='skyblue',  edgecolor='black')
plt.title('histogram')
plt.xlabel('Random values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generate random data for the histogram
data = np.random.randn(1000)

# Plotting a basic histogram
plt.hist(x=data, bins=20, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

# Save and close figure before show().
# Otherwise, the image will be blank when it is displayed.
plt.savefig('myhist.png')
plt.close('myhist.png')

# Display the plot
plt.show()
plt.close()
```

```python
fig = plt.figure()

category = ['Meal', 'Beverage', 'Transportation', 'Misc']

expense = [728, 344, 141, 72]

plt.bar(category, expense, width=0.5, align='center', color='skyblue')
plt.title('Average Weekly Expense (Baht) for College Students')
plt.xlabel('Category')
plt.ylabel('Expense (Baht)')

fig.tight_layout()
fig.savefig('bar-chart.png')
fig, axes = plt.subplots(1, 2, figsize=(6, 2), dpi=100)

axes[0].bar(category, expense, width=0.5, align='center', color='skyblue')
axes[0].set_title('Average Weekly Expense (Baht) for College Students')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Expense (Baht)')

# Using plt.barh() for horizontal bar chart.
#>>> plt.barh(category, expense, height=0.5, align='center', color='skyblue')
plt.title('Average Weekly Expense (Baht) for College Students')
plt.xlabel('Category')
plt.ylabel('Expense (Baht)')
```

```python
fig = plt.figure()
year = ['1960', '1980', '2000', '2020']
male = [15.3, 24.2, 32.3, 40.2]
female = [12.3, 20.2, 27.8, 34.0]
unspecifiled = [7.3, 12.2, 12.8, 13.0]

plt.bar(year, male, align='center', color='red', alpha=0.5, label='Male')
plt.bar(year, female, align='center', color='blue', alpha=0.7, label='Female')
plt.bar(year, unspecifiled, align='center', color='green', alpha=1, label='Unspecifiled')

plt.title('Number of Never Married Adult in U.S.')
plt.xlabel('Year')
plt.ylabel('Number')

plt.legend()
plt.tight_layout()
```

```python
fig = plt.figure()
year = ['1960', '1980', '2000', '2020']
male = [15.3, 24.2, 32.3, 40.2]
female = [12.3, 20.2, 27.8, 34.0]
unspecifiled = [7.3, 12.2, 12.8, 13.0]
index = np.arange(len(year))
bwidth = 0.15
plt.bar(index-bwidth, male, width=bwidth, align='center', color='red', alpha=1.0, label='Male')
plt.bar(index, female, width=bwidth, align='center', color='blue', alpha=1.0, label='Female')
plt.bar(index+bwidth, unspecifiled, align='center', color='yellow', alpha=1, label='Unspecifiled',width=bwidth)

plt.title('Number of Never Married Adult in U.S.')
plt.xlabel('Year')
plt.ylabel('Number')
plt.xticks(index+bwidth/2, year)
plt.legend()
plt.tight_layout()
```

```python
plt.figure(figsize=(8, 10))
snack = ['Potato Chips', 'Tortilla', 'Pretzels', 'Popcorn', 'Snack Nuts']
consumption = [11.2, 8.3, 4.3, 3.8, 2.5]
pcolors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink']
plt.pie(consumption, labels=snack, autopct='%1.1f%%', startangle=140, shadow=True, colors=pcolors)
plt.title('The Amount of Snack Food Consuption (Million Pounds)')
#plt.axis('Equal')
plt.legend()
plt.tight_layout()
```
